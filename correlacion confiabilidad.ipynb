```python
import numpy as np
import matplotlib.pyplot as plt

# Parmetros para MTBF
Base_MTBF = 100  # Valor base del MTBF
preventive_index = np.linspace(0, 1, 100)  # Valores de ndice preventivo entre 0 y 1
MTBF = Base_MTBF * (1 + 0.5 * preventive_index)

# Parmetros para Precisin_Prediccin
Base_Precision = 0.7  # Base de precisin
Learning_Rate = 0.2  # Tasa de aprendizaje
mbc_effectiveness = np.linspace(0, 1, 100)  # Valores de MBC Effectiveness entre 0 y 1
Precision_Prediction = Base_Precision + (mbc_effectiveness * Learning_Rate)

# Crear grficos
plt.figure(figsize=(12, 6))

# Grfico 1: MTBF
plt.subplot(1, 2, 1)
plt.plot(preventive_index, MTBF, label='MTBF', color='blue')
plt.title('MTBF vs Mantenimiento Preventivo Index')
plt.xlabel('Mantenimiento Preventivo Index')
plt.ylabel('MTBF')
plt.grid(True)
plt.legend()

# Grfico 2: Precisin_Prediccin
plt.subplot(1, 2, 2)
plt.plot(mbc_effectiveness, Precision_Prediction, label='Precisin Prediccin', color='green')
plt.title('Precisin Prediccin vs MBC Effectiveness')
plt.xlabel('MBC Effectiveness')
plt.ylabel('Precisin Prediccin')
plt.grid(True)
plt.legend()

# Mostrar los grficos
plt.tight_layout()
plt.show()
```
